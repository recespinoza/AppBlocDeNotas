```python
# Importar librerias
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

```

```python
# Leer datos del archivo csv
df = pd.read_csv('Prediccion_Nota_Examen_Modif.csv')
df.head()
```

```python
from pandas.core.frame import DataFrame
# Limpieza de datos
# Reemplazar edad, calidad_sueno	dificultad_exam	y puntaje_exam
# Reempalzar para las columnas numericas
df["edad"]= df["edad"].fillna(df["edad"].mean())
df["puntaje_exam"]= df["puntaje_exam"].fillna(df["puntaje_exam"].mean())

# Reemplazo para las columnas textuales
df["calidad_sueno"]= df["calidad_sueno"].fillna(df["calidad_sueno"].mode())[0]
df["dificultad_exam"]= df["dificultad_exam"].fillna(df["dificultad_exam"].mode())[0]
```

```python
# Eliminación de registros duplicados
df.drop_duplicates(inplace=True)

```

```python
# Uniformar los datos
df["genero"] = df["genero"].replace({"_femenino":"femenino", "f":"femenino", "m":"masculino"})
df.head()
```

```python
# Conversión Texto a Número
gen = {"femenino":0, "masculino": 1, "otro":2}
df["genero"] = df["genero"].map(gen)

acc = {"si": 1, "no": 0}
df["acceso_internet"] = df["acceso_internet"].map(acc)

cal = {"bueno": 2, "regular": 1, "malo":0}
df["calidad_sueno"] = df["calidad_sueno"].map(cal)

dif = {"dificil": 2, "moderado": 1, "facil":0}
df["dificultad_exam"] = df["dificultad_exam"].map(dif)


```

```python
# Definir Entradas y salidas
x = df[["edad", "genero", "horas_estudio", "acceso_internet", "horas_sueno", "calidad_sueno", "dificultad_exam"]]
y = df[["puntaje_exam"]]
```

```python
# Divisiòn de los datos
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Creación arbol de decisión
modelo = DecisionTreeRegressor(max_depth = 5)
modelo.fit(x_train, y_train)
```

```python
# Coeficiente de correlación
r2 = modelo.score(x_test, y_test * 100)
print("Coeficiente de correlación: ", r2 )
```

```python
nuevo_dato = pd.DataFrame({
  "edad": [20],
  "genero": [0],
  "horas_estudio": [7.5],
  "acceso_internet": [0],
  "horas_sueno": [5],
  "calidad_sueno": [0],
  "dificultad_exam": [1]
})
```

```python
pred1 = modelo.predict(nuevo_dato)
print("La predicción del nuevo dato es:", pred1[0])
```
